```python
x =  np.arange(-5, 5, 0.5)
y1 = 2 * x ** 2
y2 = 5 * x + 30
y3 = x + 3
y4 = x + 4
y5 = x + 5

plt.plot(x,y1, label='graph1', color= 'm')
plt.plot(x,y2, label='graph2', color= 'y')
plt.plot(x,y3, 'r:')
plt.plot(x,y4, label='graph4', color='c', marker='o')
plt.plot(x,y5,'>--g')
plt.title('This is title')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()

plt.grid(True)

plt.show()
```

```python
x = [1,2,3]
y = [3,4,6]

plt.plot(x,y, marker='o')

for i, d in enumerate(y):
    plt.text(x[i], y[i]+0.03, d, ha='center')

plt.show()
```

```python
df = {'MemberId': ['A', 'B', 'C', 'D'],
'Before': [],
'After': []
}

x = df['MemberId']
y = df['Before']

plt.bar(x,y, color='b', label='윗몸일으키기')
plt.bar(x, df['스쿼트'], color='m', bottom=y, label='스쿼트')
plt.bar(x, df['매달리기'], color = 'y', bottom=df['스쿼트']+y, label='매달리기')

plt.title('멤버별 운동량 비교')
plt.xlabel('회원ID')
plt.ylabel('운동량')
plt.legend()
plt.show()
```

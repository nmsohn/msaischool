```python
import numpy as np

x = df['ID']
y = df['']

x = np.arange(len(df))
plt.bar(x, df[''], width=0.3)
plt.bar(x+0.3, df[''], width=0.3)
plt.xticks(x, df['ID'])

plt.title(' ')
plt.xlabel('ID')
plt.ylabel('')
plt.legend()
plt.show()
```

```python
# pip install seaborn

import seaborn as sns

df = sns.load_dataset('titanic')
```

```python
df.describe()
```

```python
sns.histplot(data=df, x=df['age'], kde=True, hue=df['sex'])
plt.show()
```

```python
sns.countplot(data=df, x='survived')
plt.show()
```

```python
sns.countplot(data=df, x='class', order=df['class'].value_counts().index, palette='colorblind', hue='survived')
plt.show()
```

```python
df['class'].value_counts()
```

```python
sns.countplot(data=df, x='sex', palette='husl', hue='survived')
plt.show()
```

```python
sns.boxplot(data=df, y='age', x='pclass')
plt.show()
```

```python
sns.scatterplot(data=df, x='age', y='fare', hue='pclass')
plt.show()
```

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])
corr = df_numeric.corr()
corr
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

```python
flights = sns.load_dataset('flights')
```

```python
flights_passenger = flights.pivot_table(index='month', columns='year', values='passengers', observed=False)
```

```python
sns.heatmap(flights_passenger, annot=True, fmt='.0f', linewidths=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])

plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y1+y2, color='y')
plt.bar(x, y4, bottom=y1+y2+y3, color='g')
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```
